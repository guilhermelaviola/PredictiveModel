Counting objects in an image is a task performed by computer vision.

In this example we're counting vehicles, and for that we're using cvlib, which is a computer vision library in Python.

```python
# Installing cvlib library:
pip install cvlib
```

```python
# Importing all the necessary libraries:
import cv2
import numpy as np
import matplotlib.pyplot as plt
import cvlib as cv
from cvlib.object_detection import draw_bbox
from numpy.lib.polynomial import poly
```

```python
# Importing the image that is going to be analyzed:
image = cv2.imread("vehicles.jpeg")
```

```python
# Detecting common objects in the image:
box, label, count = cv.detect_common_objects(image)
# Generating the output:
output = draw_bbox(image, box, label, count)
```

```python
# Displaying the output:
plt.imshow(output)
plt.show()
print('There are', str(label.count('car')), 'vehicles in the image.')
```
